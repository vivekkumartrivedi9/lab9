```python
sns.jointplot(x='mpg',y='horsepower',data=df,)
```

$\huge{solution \ \ of  \ \ 2.6}$

in graph it comprises three plots. Out of the three, one plot displays a bivariate graph which shows how the dependent variable(horsepower) varies with the independent variable(mpg). Another plot is placed horizontally at the top of the bivariate graph and it shows the distribution of the independent variable(mpg). The third plot is placed on the right margin of the bivariate graph with the orientation set to vertical and it shows the distribution of the dependent variable(horsepower) here it is easier to  correlate between two data set using jointplot because it represent scatter plot of both data and also univariate plot of both data set that is easier to analyze and finding relation between both data set how it related

in scatter plot only relation between these two data set is given but in joint plot together with bivariate graph also univariate graph is provided so for finding correlation between them jointplot is easier to analyze
